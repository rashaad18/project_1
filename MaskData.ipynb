```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
county_data_path = "us-counties.csv"

# Read the mouse data and the study results
county_data = pd.read_csv(county_data_path)

# Display the data table for preview
county_data.head()
```

```python
county_data.describe()
```

```python
groupby_code= county_data.groupby(["fips"])
groupby_code.sum()
```

```python
county_df= groupby_code.sum()
county_df.head()
```

```python
county_df= county_df.reset_index()
```

```python
county_df["fips"]= county_df["fips"].astype(int)
```

```python
county_df= county_df.rename(columns= {"fips": "COUNTYFP"})
```

```python
county_df.head()
```

```python
# Bring in mask use dataset
mask_data_path = "mask-use/mask-use-by-county.csv"
```

```python
# Read the mouse data and the study results
mask_data = pd.read_csv(mask_data_path)

# Display the data table for preview
mask_data.head()
```

```python
# Merge data
county_masks= pd.merge(county_df, mask_data, how= "left", on= ["COUNTYFP", "COUNTYFP"])
county_masks.head()
```

```python
county_clean= county_masks.dropna()
county_clean.head()
```
